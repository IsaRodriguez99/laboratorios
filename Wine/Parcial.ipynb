```python
#Teniendo encuenta el tipo de datos de quality obtengo el valor medio de 'winequality-red' and 'winequality-white' 
avg_quality1 = parcial1["quality"].astype("float").mean(axis=0)
avg_quality2 = parcial2["quality"].astype("float").mean(axis=0)
print("Average of quality:", avg_quality1)
print("Average of quality:", avg_quality2)
```

```python
#Aqui realizo Normalizacion de datos 
#En este caso escale las siguientes columnas y asi al normalizar las variables para que oscile entre 0 y 1 en 'winequality-red'
parcial1['chlorides'] = parcial1['chlorides']/parcial1['chlorides'].max()
parcial1['sulphates'] = parcial1['sulphates']/parcial1['sulphates'].max()
parcial1['alcohol'] = parcial1['alcohol']/parcial1['alcohol'].max()
parcial1[["chlorides","sulphates","alcohol"]].head()
```

```python
#Aqui realizo Normalizacion de datos 
#En este caso escale las siguientes columnas y asi al normalizar las variables para que oscile entre 0 y 1 en 'winequality-white'
parcial2['chlorides'] = parcial2['chlorides']/parcial2['chlorides'].max()
parcial2['sulphates'] = parcial2['sulphates']/parcial2['sulphates'].max()
parcial2['alcohol'] = parcial2['alcohol']/parcial2['alcohol'].max()
parcial2[["chlorides","sulphates","alcohol"]].head()
```

```python
from scipy import stats
```

```python
#Aca lo que hago es usar la correlacion y causalidad, en este caso la Correlaci√≥n de Pearson
pearson_coef, p_value = stats.pearsonr(parcial1['quality'], parcial1['alcohol'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```

Como conclusion puedo decir que ya que el valor de p es $<$0.001, la correlacion entre la calidad y el alcohol es estadisticamente significativa.
